```python
def find_duplicates(lst):
    # Create a dictionary to store the count of each element
    count_dict = {}
    duplicates = []

    # Iterate through the list and update the count in the dictionary
    for item in lst:
        if item in count_dict:
            count_dict[item] += 1
        else:
            count_dict[item] = 1

    # Check for elements with count greater than 1 (i.e., duplicates)
    for item, count in count_dict.items():
        if count > 1:
            duplicates.append(item)

    return duplicates


# Find duplicate entries in the list
duplicate_entries = find_duplicates(image_file_names)

# Output the duplicate entries
print("Duplicate entries:", duplicate_entries)
```
