```python
import os
from pathlib import Path
import pickle
```

```python
GDRIVE_EXPW_RACEDS_BASE_PATH = Path("/content/gdrive/MyDrive/CV_FER/dataset")
EXPW_RACEDS_EXTRACT_PATH = Path(GDRIVE_EXPW_RACEDS_BASE_PATH,'RaceTest_EXPW')
EXPW_RACEDS_ANNOTATED_PATH = Path(EXPW_RACEDS_EXTRACT_PATH,'EXPW_annotated')

labels_map={"0":"angry",
            "1":"disgust",
            "2":"fear",
            "3":"happy",
            "4":"sad",
            "5":"surprise",
            "6":"neutral"}
labels_map_reverse = {v: k for k, v in labels_map.items()}

emotion_list = list(labels_map.values())
```

```python
#@title Storing the image files in a list
image_file_names = []
image_label_dict = {}
for dir in os.listdir(EXPW_RACEDS_ANNOTATED_PATH):
  dir_lower = str(dir).strip().lower()
  if dir_lower in emotion_list:
    race_list = [item for item in os.listdir(os.path.join(EXPW_RACEDS_ANNOTATED_PATH,dir)) if not os.path.isdir(item)]
    # print(dir,":", race_list)
    for race in race_list:
      image_files = [ item for item in os.listdir(os.path.join(EXPW_RACEDS_ANNOTATED_PATH,dir,race))]
      # print(dir_lower,":",race,":",image_files)
      image_file_names.extend(image_files)
      for image_file in image_files:
        image_label_dict[image_file] = int(labels_map_reverse[dir_lower])

print(len(image_file_names),"::",image_file_names)
```

```python
for k, v in image_label_dict.items():
  print(k,"::::::::::::",v)
```

```python
#@title Storing the list on google drive
file_path_pkl = os.path.join(GDRIVE_EXPW_RACEDS_BASE_PATH,"expw_raceds_image_list.pkl")
dict_path_pkl = os.path.join(GDRIVE_EXPW_RACEDS_BASE_PATH,"expw_raceds_image_dict.pkl")

print(file_path_pkl , dict_path_pkl)

# Serialize  list using Pickle and save it to a file in Google Drive
with open(file_path_pkl, 'wb') as f:
    pickle.dump(image_file_names, f)

# Serialize  list using Pickle and save it to a file in Google Drive
with open(dict_path_pkl, 'wb') as f:
    pickle.dump(image_label_dict, f)
```
